```python
import spacy
from rich.console import Console
from rich.text import Text

# Charger le modèle de langue anglais
nlp = spacy.load("en_core_web_sm")

# Charger le texte depuis le fichier data.txt
with open('data.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# Analyser le texte pour détecter les entités
doc = nlp(text)

# Préparation pour afficher le texte avec les noms/prénoms en couleur
console = Console(record=True, width=100)
formatted_text = Text()

last_end = 0

# Couleurs par type d’entité
color_map = {
    "PERSON": "red"
}

# Boucle pour ajouter les entités en couleur
for ent in doc.ents:
    if ent.label_ == "PERSON":  # Identifier uniquement les entités de type "PERSON"
        # Ajouter du texte normal entre les entités
        formatted_text.append(text[last_end:ent.start_char])
        # Ajouter l'entité de type PERSON en couleur
        entity_color = color_map.get(ent.label_, "yellow")
        formatted_text.append(ent.text, style=f"bold {entity_color}")
        last_end = ent.end_char

# Ajouter le texte restant après la dernière entité
formatted_text.append(text[last_end:])

# Affichage du texte avec `rich` dans la console
console.print(formatted_text)

# Exporter le texte en format HTML
html_output = console.export_html(inline_styles=True)

# Enregistrer dans un fichier HTML
with open("ner_output.html", "w", encoding="utf-8") as html_file:
    html_file.write(html_output)

print("Le fichier HTML 'ner_output.html' a été généré avec les noms et prénoms en couleur.")
```

```python
import spacy
import json
from sklearn.metrics import precision_recall_fscore_support

# Load SpaCy English model
nlp = spacy.load("en_core_web_sm")

# Load the annotated data from Label Studio JSON
with open('label_studio_annotations.json', 'r', encoding='utf-8') as file:
    annotations = json.load(file)

# Prepare lists for true labels and predicted labels
true_entities = []
predicted_entities = []

# Extract entities from the Label Studio annotations
for item in annotations:
    for annotation in item['annotations'][0]['result']:
        if annotation['type'] == 'labels' and annotation['value']['labels'][0] == 'PERSON':
            start = annotation['value']['start']
            end = annotation['value']['end']
            entity_text = item['text'][start:end]
            true_entities.append((entity_text, "PERSON"))

# Process the text with the SpaCy model and extract predicted entities
with open('data.txt', 'r', encoding='utf-8') as file:
    text = file.read()

doc = nlp(text)
for ent in doc.ents:
    if ent.label_ == "PERSON":
        predicted_entities.append((ent.text, ent.label_))

# Convert true and predicted entities into sets for comparison
true_set = set(true_entities)
pred_set = set(predicted_entities)

# Calculate precision, recall, and F1 score
tp = len(true_set & pred_set)
fp = len(pred_set - true_set)
fn = len(true_set - pred_set)

precision = tp / (tp + fp) if (tp + fp) > 0 else 0
recall = tp / (tp + fn) if (tp + fn) > 0 else 0
f1_score = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0

# Display metrics
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1_score:.2f}")
```

```python
import spacy
import json
from spacy.training import Example

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Step 1: Load and predict entities from data_corrected.txt
with open('data_corrected.txt', 'r', encoding='utf-8') as f:
    text = f.read()

# Get model predictions on the text
doc = nlp(text)
predicted_entities = [(ent.start_char, ent.end_char, ent.label_) for ent in doc.ents if ent.label_ == "PERSON"]

# Step 2: Load ground truth annotations from label_studio_annotations.json
with open('label_studio_annotations.json', 'r', encoding='utf-8') as f:
    annotations = json.load(f)

# Prepare evaluation examples
examples = []
for item in annotations:
    text = item['data']['text']
    entities = []
    for result in item['annotations'][0]['result']:
        if result['value']['labels'][0] == "PER":
            start = result['value']['start']
            end = result['value']['end']
            label = "PERSON"
            entities.append((start, end, label))
    # Create a Doc object with the text
    doc = nlp.make_doc(text)
    # Create Example object with ground truth entities
    example = Example.from_dict(doc, {"entities": entities})
    examples.append(example)

# Step 3: Evaluate using spaCy's evaluate function
evaluation_results = nlp.evaluate(examples)

# Extract precision, recall, and F1 score
precision = evaluation_results['ents_p']*100
recall = evaluation_results['ents_r'] *100
f1_score = evaluation_results['ents_f']*100

# Print the results
print(f"Precision (ents_p): {precision:.2f}")
print(f"Recall (ents_r): {recall:.2f}")
print(f"F1 Score (ents_f): {f1_score:.2f}")
```

```python
import json
import spacy
from spacy.tokens import DocBin

# Load the annotated data
with open('label_studio_annotations.json', 'r', encoding='utf-8') as f:
    annotations = json.load(f)

# Initialize an empty DocBin for storing training examples
db = DocBin()

# Load the spaCy model to create Doc objects
nlp = spacy.blank("en")  # Start with a blank model

for item in annotations:
    text = item['data']['text']
    entities = []
    for result in item['annotations'][0]['result']:
        if result['value']['labels'][0] == "PER":
            start = result['value']['start']
            end = result['value']['end']
            entities.append((start, end, "PERSON"))
    # Create a Doc object with entities
    doc = nlp.make_doc(text)
    spans = [doc.char_span(start, end, label=label) for start, end, label in entities]
    doc.ents = [span for span in spans if span is not None]
    db.add(doc)

# Save the training data
db.to_disk("training_data.spacy")
```

```python
import spacy
import json
from spacy.training import Example
from rich.console import Console
from rich.text import Text

# Load the fine-tuned model
nlp_fine_tuned = spacy.load("./output/model-best")

# Load the text from ticket2.txt
with open('ticket2.txt', 'r', encoding='utf-8') as f:
    text = f.read()

# Use the fine-tuned model to predict entities in the text
doc = nlp_fine_tuned(text)

# Initialize the console for displaying the text with detected entities
console = Console()
highlighted_text = Text(text)

# Highlight detected "PERSON" entities in red bold
for ent in doc.ents:
    if ent.label_ == "PERSON":
        highlighted_text.stylize("bold red", ent.start_char, ent.end_char)

# Display the highlighted text
console.print(highlighted_text)

# Load ground truth annotations from ticket2.json
with open('ticket2.json', 'r', encoding='utf-8') as f:
    annotations = json.load(f)

# Prepare evaluation examples based on ground truth annotations
examples = []
for item in annotations:
    annotated_text = item['data']['text']
    entities = [(result['value']['start'], result['value']['end'], "PERSON") 
                for result in item['annotations'][0]['result'] 
                if result['value']['labels'][0] == "PER"]
    
    # Create a Doc object with the annotated text
    doc = nlp_fine_tuned.make_doc(annotated_text)
    
    # Create an Example object with the Doc and ground truth entities
    example = Example.from_dict(doc, {"entities": entities})
    examples.append(example)

# Evaluate the fine-tuned model on the prepared examples
evaluation_results = nlp_fine_tuned.evaluate(examples)

# Extract and print precision, recall, and F1 score
precision = evaluation_results['ents_p'] * 100
recall = evaluation_results['ents_r'] * 100
f1_score = evaluation_results['ents_f'] * 100

print("\nEvaluation Metrics for Fine-Tuned Model:")
print(f"Precision (ents_p): {precision:.2f}%")
print(f"Recall (ents_r): {recall:.2f}%")
print(f"F1 Score (ents_f): {f1_score:.2f}%")
```
