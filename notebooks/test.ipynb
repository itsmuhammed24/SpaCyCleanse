```python
from presidio_analyzer import AnalyzerEngine
from termcolor import colored

# Initialisation du moteur d'analyse
analyzer = AnalyzerEngine()

# Chargement du texte depuis le fichier data.txt
with open('data.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# Phase d'analyse pour identifier les entités sensibles
entities = analyzer.analyze(text=text, entities=["PERSON", "ORGANIZATION", "PHONE_NUMBER", "EMAIL_ADDRESS"], language="en")

# Tri des entités par position de départ pour éviter des conflits lors de l'affichage coloré
entities = sorted(entities, key=lambda x: x.start)

# Couleurs par catégorie d'entités
color_map = {
    "PERSON": "red",
    "ORGANIZATION": "blue",
    "PHONE_NUMBER": "green",
    "EMAIL_ADDRESS": "magenta"
}

# Variable pour stocker le texte coloré
colored_text = ""
last_end = 0

# Construction du texte coloré avec les entités identifiées
for entity in entities:
    # Ajout du texte non identifié entre les entités
    colored_text += text[last_end:entity.start]
    # Texte de l'entité identifié en couleur spécifique
    entity_color = color_map.get(entity.entity_type, "yellow")  # 'yellow' par défaut pour toute entité non définie
    colored_entity = colored(text[entity.start:entity.end], entity_color, attrs=['bold'])
    colored_text += colored_entity
    last_end = entity.end

# Ajout du texte restant après la dernière entité
colored_text += text[last_end:]

# Affichage du texte coloré
print("Texte avec entités en couleurs :")
print(colored_text)
```

```python
from presidio_analyzer import AnalyzerEngine
from rich.console import Console
from rich.text import Text

# Initialisation du moteur d'analyse
analyzer = AnalyzerEngine()

# Chargement du texte depuis le fichier data.txt
with open('data.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# Phase d'analyse pour identifier les entités sensibles (sans "ORGANIZATION")
entities = analyzer.analyze(text=text, entities=["PERSON", "PHONE_NUMBER", "EMAIL_ADDRESS"], language="en")

# Tri des entités par position de départ pour éviter des conflits lors de l'affichage coloré
entities = sorted(entities, key=lambda x: x.start)

# Couleurs par catégorie d'entités
color_map = {
    "PERSON": "red",
    "PHONE_NUMBER": "green",
    "EMAIL_ADDRESS": "magenta"
}

# Préparation du texte coloré
console = Console(record=True, width=100)
formatted_text = Text()

last_end = 0

# Construction du texte coloré avec les entités identifiées
for entity in entities:
    # Ajouter du texte normal entre les entités
    formatted_text.append(text[last_end:entity.start])
    # Entité en couleur
    entity_color = color_map.get(entity.entity_type, "yellow")  # 'yellow' par défaut
    formatted_text.append(text[entity.start:entity.end], style=f"bold {entity_color}")
    last_end = entity.end

# Ajouter le texte restant après la dernière entité
formatted_text.append(text[last_end:])

# Affichage du texte avec `rich` dans la console
console.print(formatted_text)

# Exporter la console au format HTML sans thème pour éviter les erreurs de compatibilité
console.save_html("output.html")
```
